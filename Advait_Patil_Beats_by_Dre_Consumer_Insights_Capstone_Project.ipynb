Load the data below and process cleaning here.

```python
#For data cleaning, we had to import numpy and pandas
import pandas as pd
import numpy as np

#We use pandas to read the csv contents and convert it into a dataframe
df = pd.read_csv('/content/sample_data/speaker_review_data.csv')

#We use the head and tail methods to get an idea of what the data looks like
df.head()
df.tail()

#The info method gives us an idea of the various datatypes found in the dataframe
df.info()

#The shape method provides the total number of rows and columns
df.shape

#Any proper data cleaning requires taking care of null values. The code below gives us the total number of null values
df.isna().sum()

#Using the dropna() method gets rid of any null values in the dataframe
df = df.dropna()
df.isna().sum()

#The describe method tells us the aggregate data of the dataframe
df.describe()

#Outliers are another attribute that requires specific attention. I used the IQR method to determine the total number of outliers
q1 = df['helpful_count'].quantile(0.25)
q3 = df['helpful_count'].quantile(0.75)
iqr = q3 - q1

threshold = 1.5
outliers = df[(df['helpful_count'] < q1 - (threshold * iqr)) | (df['helpful_count'] > q3 + (threshold * iqr))]
```

```python
outliers
```

Because the outliers deal with the number of people that found a certain review helpful, I decided not to remove these reviews as they were valuable to the integrity of the data

## 6. Sentiment Analysis

### 6.1 Methodology
Explain the sentiment analysis techniques used (e.g., TextBlob, NLTK) and provide code snippets.

For my sentiment analysis, I used the TextBlob library to determine the polarity of various reviews and the WordCloud library to visualize reviews of similar sentiment. This data was later visualized to determine aggregate values and key words used in each review

```python
from textblob import TextBlob
positive_reviews = []
neutral_reviews = []
negative_reviews = []

for review in df['content']:
  nlp_analysis = TextBlob(review)
  polarity = nlp_analysis.sentiment.polarity
  if polarity > 0:
    positive_reviews.append([review,polarity])
  elif polarity == 0:
    neutral_reviews.append([review,polarity])
  else:
    negative_reviews.append([review,polarity])
```

```python
positive_reviews_df = pd.DataFrame(positive_reviews,columns=['review','polarity'])
```

```python
neutral_reviews_df = pd.DataFrame(neutral_reviews,columns=['review','polarity'])
negative_reviews_df = pd.DataFrame(negative_reviews,columns=['review','polarity'])
```

```python
from wordcloud import WordCloud
```

```python
positive_text = positive_reviews_df['review'].str.cat(sep='')

wordcloud = WordCloud(collocations=False).generate(positive_text)

plt.figure(figsize=(8, 4))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
negative_text = negative_reviews_df['review'].str.cat(sep='')

wordcloud = WordCloud(collocations=False).generate(negative_text)

plt.figure(figsize=(8,4))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```
